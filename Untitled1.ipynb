```python
import matplotlib.pyplot as plt

class WeatherData:
    def __init__(self, temperature, humidity, rainfall, pm25, pm10, no2):
        self.temperature = temperature
        self.humidity = humidity
        self.rainfall = rainfall
        self.pm25 = pm25
        self.pm10 = pm10
        self.no2 = no2

    def display_statistics(self):
        data = {
            "Temperature": self.temperature,
            "Humidity": self.humidity,
            "Rainfall": self.rainfall,
            "PM2.5": self.pm25,
            "PM10": self.pm10,
            "NO2": self.no2
        }

        for key, values in data.items():
            print(f"\n{key}")
            print("Minimum:", min(values))
            print("Maximum:", max(values))
            print("Average:", sum(values)/len(values))

    def check_thresholds(self):
        for value in self.pm25:
            if value > 50:
                print("WARNING: PM2.5 exceeds WHO limit")

        for value in self.pm10:
            if value > 100:
                print("WARNING: PM10 exceeds standard")

    def plot_data(self):
        plt.plot(self.pm25, label="PM2.5")
        plt.plot(self.pm10, label="PM10")
        plt.title("Air Quality Monitoring")
        plt.xlabel("Month")
        plt.ylabel("Concentration")
        plt.legend()
        plt.show()

try:
    temperature = [32,34,31,35,33]
    humidity = [70,75,72,78,80]
    rainfall = [20,10,30,15,25]
    pm25 = [45,60,40,70,55]
    pm10 = [80,110,95,120,90]
    no2 = [25,30,35,40,28]

    weather = WeatherData(
        temperature,
        humidity,
        rainfall,
        pm25,
        pm10,
        no2
    )

    weather.display_statistics()
    weather.check_thresholds()
    weather.plot_data()

except Exception as e:
    print("Error:", e)
```
